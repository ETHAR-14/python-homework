```python
import numpy as np

A = np.array([[5,8,2,7],
              [1,9,4,3],
              [6,0,11,10],
              [12,13,14,15]])

print(A[1])
print(A[:,2])
print(A[1:3,1:3])
```

```python
scores = np.array([[70,80,90],
                   [60,75,85],
                   [88,92,79],
                   [55,65,72]])

scores[:,1] += 5
print(scores)

scores[0] = [100,100,100]
print(scores)

scores[:,2] = 0
print(scores)
```

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

v = np.array([2,4])

print(A * B)
print(np.dot(A,B))
print(A * v)
```

```python
np.random.seed(0)

arr = np.random.randint(-5,6,(4,4))

values, counts = np.unique(arr, return_counts=True)
print(values)
print(counts)

print(np.unique(arr, axis=0))
```

```python
a = np.array([[1,0,0],
              [1,0,0],
              [2,3,4]])

print(np.unique(a, axis=0))
```

```python
x = np.array([2,0,1,5,4,1,9])

print(np.argsort(x))
print(x[np.argsort(x)])
```
