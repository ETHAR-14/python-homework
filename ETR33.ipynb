```python
print("1- Addition")
print("2- Subtraction")
print("3- Multiplication")
print("4- Division")

ch = input("enter choose number: ")

fn = float(input("enter first number: "))
sn = float(input("enter second number: "))

match ch:
    case "1":
        print(fn, "+", sn, "=", fn+sn)

    case "2":
        print(fn, "-", sn, "=", fn-sn)

    case "3":
        print(fn, "*", sn, "=", fn*sn)

    case "4":
        print(fn, "/", sn, "=", fn/sn)

    case _:
        print("invalid")
```

```python
score = int(input("Enter student score: "))

if score >= 60:
    status = "Passed"
else:
    status = "Failed"

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

match grade:
    case "A":
        print("Excellent")
    case "B":
        print("Very Good")
    case "C":
        print("Good")
    case "D":
        print("Pass")
    case "F":
        print("Fail")
```
