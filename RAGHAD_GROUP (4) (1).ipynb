```python

# Student Management System

# رغد جايز محمد الفقيري
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Invalid age")


class Student(Person):
    def __init__(self, name, age, student_id, grade):
        super().__init__(name, age)
        self.student_id = student_id
        self.__grade = grade

    def get_grade(self):
        return self.__grade

    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade
        else:
            print("Invalid grade")


# --------- Data Storage ---------
students = []
FILE_NAME = "data.txt"


def find_student(student_id):
    for s in students:
        if s.student_id == student_id:
            return s
    return None


# ايثار ضاحي
def add_student():
    name = input("Enter name: ").strip()
    if name == "":
        print("Invalid name")
        return

    try:
        age = int(input("Enter age: "))
        if age <= 0:
            print("Invalid age")
            return
    except:
        print("Age must be a number")
        return

    student_id = input("Enter ID: ").strip()
    if student_id == "" or find_student(student_id):
        print("Invalid or duplicate ID")
        return

    try:
        grade = float(input("Enter grade: "))
        if grade < 0 or grade > 100:
            print("Invalid grade")
            return
    except:
        print("Grade must be a number")
        return

    students.append(Student(name, age, student_id, grade))
    save_to_file()
    print("Student added successfully!")


def delete_student():
    student_id = input("Enter ID to delete: ")
    student = find_student(student_id)

    if student:
        students.remove(student)
        save_to_file()
        print("Deleted successfully!")
    else:
        print("Student not found!")


def update_student():
    student_id = input("Enter ID to update: ")
    student = find_student(student_id)

    if student:
        new_name = input("New name: ").strip()
        if new_name:
            student.name = new_name

        try:
            new_age = input("New age: ")
            if new_age:
                student.set_age(int(new_age))
        except:
            print("Invalid age")

        try:
            new_grade = input("New grade: ")
            if new_grade:
                student.set_grade(float(new_grade))
        except:
            print("Invalid grade")

        save_to_file()
        print("Updated successfully!")
    else:
        print("Student not found!")


def display_students():
    if not students:
        print("No students found")
        return

    for s in students:
        print("-----")
        print("Name:", s.name)
        print("Age:", s.get_age())
        print("ID:", s.student_id)
        print("Grade:", s.get_grade())


# اجوان
def save_to_file():
    with open(FILE_NAME, "w") as file:
        for s in students:
            file.write(f"{s.name},{s.get_age()},{s.student_id},{s.get_grade()}\n")


def load_from_file():
    try:
        with open(FILE_NAME, "r") as file:
            for line in file:
                data = line.strip().split(",")

                name = data[0]
                age = int(data[1])
                student_id = data[2]
                grade = float(data[3])

                students.append(Student(name, age, student_id, grade))

    except FileNotFoundError:
        print("No previous data found, starting fresh.")


# نورا
def menu():
    while True:
        print("\nStudent System ")
        print("1. Add Student")
        print("2. Delete Student")
        print("3. Update Student")
        print("4. Display Students")
        print("5. Save & Exit")

        choice = input("Enter your choice: ")
        if choice == "1":
            add_student()

        elif choice == "2":
            delete_student()

        elif choice == "3":
            update_student()

        elif choice == "4":
            display_students()

        elif choice == "5":
            save_to_file()
            print("Goodbye!")
            break

        else:
            print("Invalid choice")



load_from_file()
menu()
```
