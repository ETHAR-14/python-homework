```python
import os

folder = "lab_data"

if not os.path.exists(folder):
    os.mkdir(folder)

subfolders = ["input", "output", "logs"]
for sub in subfolders:
    path = os.path.join(folder, sub)
    if not os.path.exists(path):
        os.mkdir(path)

for item in os.listdir(folder):
    full_path = os.path.join(folder, item)
    if os.path.isdir(full_path):
        print(item, "Folder")
    else:
        print(item, "File")

os.rename(os.path.join(folder, "logs"), os.path.join(folder, "history"))
 
 
```

```python
import os

notes_path = os.path.join("lab_data", "input", "notes.txt")

with open(notes_path, "w") as f:
    f.write("Line 1\nLine 2\nLine 3\nLine 4\nLine 5\n")

with open(notes_path, "r") as f:
    lines = f.readlines()
    print("Number of lines:", len(lines))

summary_path = os.path.join("lab_data", "output", "summary.txt")

with open(summary_path, "w") as f:
    f.writelines(lines)

```

```python
import os

history_path = os.path.join("lab_data", "history")

if os.path.exists(history_path):
    if len(os.listdir(history_path)) == 0:
        os.rmdir(history_path)
        print("Deleted history folder")
    else:
        print("Folder is not empty")
else:
    print("Folder not found")
```
