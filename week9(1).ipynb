```python
import numpy as np

A = np.arange(1, 21).reshape(4, 5)
print(A)
print(A.dtype)
print(A.ndim)
print(A.shape)
print(A.size)
print(A.itemsize)

B = A.reshape(5, 4)
print(B)
```

```python
a = np.array([[0,0,0],
              [10,10,10],
              [20,20,20],
              [30,30,30]])

b = np.array([0,1,2])

c = a + b
print(c)

mask = (c >= 22)
print(mask)

print(np.sum(c >= 22))
```

```python
grades = np.array([[87,96,70],
                   [100,87,90],
                   [94,77,90],
                   [100,81,82]])

print(np.sum(grades))
print(np.min(grades))
print(np.max(grades))
print(np.mean(grades))
print(np.std(grades))
print(np.var(grades))

print(np.mean(grades, axis=0))
print(np.mean(grades, axis=1))
print(np.max(grades, axis=1))
```

```python
arr = np.array([[2,4,6],
                [1,3,5],
                [7,9,11]])

print(np.sum(arr, axis=1))
print(np.sum(arr, axis=0))
print(np.mean(arr))
```

```python
A = np.array([[10,20,30],
              [40,50,60]])

B = np.array([1,2,3])

print(A + B)
print(A * 2)
print(A - 5)
```

```python
grades = np.array([[70,85,90],
                   [88,76,95],
                   [60,80,72]])

print(np.argmax(grades))
print(np.unravel_index(np.argmax(grades), grades.shape))

print(np.argmax(grades, axis=1))
print(np.argmin(grades, axis=0))
```
